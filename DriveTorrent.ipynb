```python
# @title Torrent to G drive
# @markdown This uses Aria2c, which avoids the "libtorrent" Python version errors completely.
magnet_link = "" # @param {type:"string"}
output_path = "/content/drive/My Drive/Downloads" # @param {type:"string"}

import os
import sys
from google.colab import drive

# 1. Mount Google Drive
if not os.path.exists('/content/drive'):
    print("Mounting Google Drive...")
    drive.mount('/content/drive')

# Ensure output directory exists
if not os.path.exists(output_path):
    os.makedirs(output_path)

# 2. Install Aria2 (The System Tool)
# We use standard apt which is always compatible with the OS kernel
if os.system("which aria2c") != 0:
    print("Installing Aria2c (Reliable Torrent Client)...")
    !apt install -y aria2 > /dev/null

# 3. Clean and Validate Link
if not magnet_link:
    print(" Error: Please enter a magnet link.")
else:
    print(f" Starting Download to: {output_path}")
    print("âš¡ Tracker connection established. Please wait for metadata...")
    print("-" * 60)

    # 4. Run Download
    # --seed-time=0: Stop seeding immediately after download
    # --summary-interval=0: Keeps the log clean
    # -d: Output directory
    cmd = f'aria2c --seed-time=0 -d "{output_path}" "{magnet_link}"'

    # Executing the command directly shows the native high-speed progress bar
    !$cmd

    print("\n" + "-" * 60)
    print(" Download Finished! Check your Google Drive.")
```
